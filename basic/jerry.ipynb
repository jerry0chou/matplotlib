```python
import numpy as np
import matplotlib.pyplot as plt
N = 100
x = np.random.randn(N)
y = -x + np.random.randn(N) * 0.2
plt.scatter(x, y,  # 坐标
            s=100,  # 面积
            c='g',  # 颜色
            alpha=0.3  # 透明度 可以判断 哪些点是集中的
            )
plt.show()
```
